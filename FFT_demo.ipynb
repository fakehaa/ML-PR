```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = '/content/messi.jpg'
img = cv2.imread(path, 0)
# show the images for reference
```

```python

plt.figure(figsize=(10,5))
plt.imshow(img)
plt.title('origin')
plt.show()
```

```python
f = np.fft.fft2(img)
fs = np.fft.fftshift(f)
lfs = 15 * np.log(1 + np.abs(fs))
lfs = np.uint8(lfs)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(lfs)
plt.title('dft')
plt.show()
```

```python
ishift_img = np.fft.ifftshift(fs)
ifft_img = np.fft.ifft2(ishift_img)
abs_img = np.abs(ifft_img)
abs_img =np.uint8(abs_img)
```

```python
plt.figure(figsize=(10,5))
plt.imshow(abs_img )
plt.title('idft')
plt.show()
```
